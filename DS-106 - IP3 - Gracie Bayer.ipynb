#### 5.	Remove all the symbols ("$") from the 'Price' column and convert and update it to  64-bit int numbers using astype(int64) (Hint: Look for string functions like replace/strip.)

```python
car_sales["Price"]=car_sales["Price"].replace("[$,]", "", regex=True).astype('int64')
car_sales.head()
```

#### 6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column. Look up the cumsum() function.

```python
car_sales['Total Sales']=car_sales['Price'].cumsum()
car_sales.head()
```

#### 7. Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out date_range function and keep the periods=len(car_sales).

```python
car_sales["Sale Date"]=pd.date_range("2/22/2022", periods=len(car_sales))
type(car_sales["Price"][0])
car_sales.head()
```

#### 8. Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.)

```python
car_sales.plot(x="Price", y="Total Sales")
plt.title('Price vs Total Sales', size='large')
plt.xlabel('Price', color='red')
plt.ylabel('Total Sales', color='red')
plt.show()
```

#### 9. Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
plt.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
plt.title('Odometer (KM) vs Price', size='large')
plt.xlabel('Odometer (KM)', color='red')
plt.ylabel('Price', color='red')
plt.show()
```

#### 10. Create a matplotlib bar graph of the 'Make' vs 'Odometer (KM)' columns.

```python
plt.bar(car_sales['Odometer (KM)'], car_sales['Make'], color = "blue")
plt.xlabel("Odometer (KM)", color='red')
plt.ylabel("Make", color='red')
plt.title("Make vs Odometer (KM)", size='large')
plt.show()
```

#### 11. Create a histogram of the 'Make' column.

```python
plt.hist(car_sales['Make'], facecolor='lightgreen', edgecolor='blue', bins=10, cumulative=True)
plt.title('Histogram of Car Make', size='large')
plt.show()
```

#### 12. Create a subplot of multiple plots (steps 8, 9, 10, & 11).

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(2,2,figsize=(12,10))

ax1.plot(car_sales["Price"],car_sales["Total Sales"])
ax1.set_title('Price vs Total Sales', size='large')
ax1.set_xlabel('Price', color='red')
ax1.set_ylabel('Total Sales', color='red')
car_sales['Price']=car_sales['Price'].astype(int)

ax2.scatter(car_sales['Odometer (KM)'],car_sales['Price'])
ax2.set_title('Odometer (KM) vs Price', size='large')
ax2.set_xlabel('Odometer (KM)', color='red')
ax2.set_ylabel('Price', color='red')

ax3.bar(car_sales['Odometer (KM)'], car_sales['Make'])
ax3.set_title('Make vs Odometer (KM)', size='large')
ax3.set_xlabel('Odometer (KM)', color='red')
ax3.set_ylabel('Make', color='red')

ax4.hist(car_sales['Make'], facecolor='lightgreen', edgecolor='blue', bins=10, cumulative=True)
ax4.set_title('Make', size='large')

plt.show()
```

#### 16. Find out Pearson correlation between 'Total Sales', 'Price', 'Odometer (KM)'. Comment on the correlation.

```python
pc=car_sales[['Total Sales','Price','Odometer (KM)']].corr(method ='pearson')
ax=sns.heatmap(pc, annot=True)
```
